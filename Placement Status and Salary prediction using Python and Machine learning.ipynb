# Outlier Analysis


```python
sns.boxplot(y='degree_p', data=df)
```

```python
percentile25=df['degree_p'].quantile(0.25)
percentile75=df['degree_p'].quantile(0.75)
```

```python
iqr=percentile75-percentile25
```

```python
upper_limit=percentile75+(1.5*iqr)
lower_limit=percentile25-(1.5*iqr)
```

```python
upper_limit=percentile75+(1.5*iqr)
lower_limit=percentile25-(1.5*iqr)
df=df[df['degree_p']<upper_limit]
df=df[df['degree_p']>lower_limit]
```

```python
df
```

```python
sns.boxplot(y='degree_p', data=df)
```

```python
col=['ssc_p','hsc_p','degree_p','etest_p','mba_p','salary']
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
```

```python
for c in col:
  percentile25=df[c].quantile(0.25)
  percentile75=df[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  df=df[df[c]<upper_limit]
  df=df[df[c]>lower_limit]
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
new_df=df.copy()
```

```python
kmeans=KMeans(3)
```

```python
kmeans.fit(new_df)
```

```python
clusters=kmeans.predict(new_df)
```

```python
new_df['Cluster']=clusters
```

```python
new_df
```

```python
kmeans.inertia_
```

```python
new_df.reset_index()
```

```python
inertia=[]
for i in range(1,11):
  kmeans=KMeans(i)
  kmeans.fit(new_df)
  inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
clusters=range(1,11)
```

```python
plt.figure(figsize=(6,6))
plt.plot(clusters,inertia,marker='o')
plt.title("Elbow Curve")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
```

# PCA

```python
X=df.reset_index().drop(columns=['status','salary','index'],axis=1)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_trans=StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
prinicipal_components=pca.fit_transform(X_trans)
```

```python
prinicipal_components
```

```python
pca_df=pd.DataFrame(data=prinicipal_components,columns=['PC1','PC2'])
```

```python
final_df=pd.concat([pca_df,df.reset_index()[['status','salary']]],axis=1)
```
